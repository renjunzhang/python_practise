1、布尔类型和比较计算符
    
    比较计算符用于比较两个值，结果是布尔类型（True 或 False）。

```python
bool_1 = True
bool_2 = False
print(type(bool_1))  # <class 'bool'>
print(bool_1 and bool_2)  # 与运算，结果为 False

# 比较计算符
compare_1 = 10
compare_2 = 20
print(compare_1 != compare_2)  # 不等于，结果为 True
print(f"两个数是否相等: {compare_1 == compare_2}")  # 等于，结果为 False
```

2、if 判断语句

```python
age = 22
if age >=18:
    print("成年人")
else:
    print("未成年人")
print("成年人" if age >= 18 else "未成年人")  # 三元表达式
```

3、if elif else 语句

    多个条件判断时，可以使用 if、elif 和 else 语句。
    当满足第一个条件时，执行对应的代码块；如果第一个条件不满足，则检查下一个条件，直到找到满足的条件或所有条件都不满足。

```python
print("welcome to Macao casino")
your_salary = int(input("请输入您的薪水: "))
your_family_members = int(input("请输入您的家庭成员数量: "))
# 多条件判断
if your_salary > 5000:
    print("您属于高收入群体,欢迎您来 Macao 赌场！")
elif your_family_members > 5:
    print("您有个大家庭,欢迎您来 Macao 赌场！")
else:
    print("您属于低收入群体,请您滚！")
```

4、判断语句的嵌套

```python
age = 25
year = 3
level = 5
if age >=18:
    print("成年人")
    if age < 30:
        print("你的年龄达标了，在18到30岁之间")
        if year >= 3:
            print(f"你工作满{year}年了，可以申请信用卡")
        elif level >= 5:
            print("你的级别达标，可以申请信用卡")
        else:
            print("你不符合申请信用卡的条件")
    else:
        print("你的年龄不符合申请信用卡的条件")
else:
    print("未成年人")
```

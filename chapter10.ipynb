魔术方法

```python
#字符串方法
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"Student类对象，name:{self.name},age:{self.age}"
stu = Student("张三",18)
print(stu)
#lt方法
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __lt__(self, other):
        return self.age<other.age
stu1 = Student("张三",18)
stu2 = Student("李四",19)
print(stu1<stu2)
#le方法
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __le__(self, other):
        return self.age<other.age
stu1 = Student("张三",20)
stu2 = Student("李四",19)
stu3 = Student("王五",19)
print(stu2<=stu3)
#eq方法
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __eq__(self, other):
        return self.age<other.age
stu1 = Student("张三",20)
stu2 = Student("李四",19)
stu3 = Student("王五",19)
print(stu2==stu3)
```

封装

    封装是指将数据和方法封装在一个类中，限制外部对数据的直接访问。
    在Python中，使用单下划线前缀（_）表示保护成员，双下划线前缀（__）表示私有成员。
    通过属性装饰器（@property）可以控制对属性的访问。

    私有成员变量和方法只能在类的内部访问，外部无法直接访问。

```python
"""
    演示面向对象封装思想中私有成员的使用"l n1
    没有return会返回 None
"""
#定义一个类，内含私有成员变量和私有成员方法
class Phone:
    __current_voltage = 1 #当前手机运行电压
    def __keep_single_core(self):
        print("让CPU单核运行")
        # 没有return会返回 None
    def call_by_5G(self):
        if self.__current_voltage >=1:
            print("5G通话已开启")
        else:
            self.__keep_single_core()
            print("电量不足无法使用5G，已切换为单核")
phone = Phone()
# phone._keep_single_core()
# phone.call_by_5G()
print(phone._Phone__current_voltage) # 访问私有成员变量
try:
    print(phone.__current_voltage) # 直接访问私有成员变量会报错
except AttributeError as e:
    print(e)

print(phone._Phone__keep_single_core())
try:
    phone.__keep_single_core()  # 直接访问私有成员方法会报错
except AttributeError as e:
    print(e)

print(phone.call_by_5G())  # 调用公共方法，间接访问私有成员方法
try:
    phone.call_by_5G()  # 调用公共方法，间接访问私有成员方法
except AttributeError as e:
    print(e)
```

继承

    继承是指一个类可以继承另一个类的属性和方法。
    在Python中，使用括号表示继承，例如：class ChildClass(ParentClass)。
    子类可以重写父类的方法，也可以添加新的方法和属性。

```python
"""
演示面向对象:继承的基础语法
"""


# 演示单继承
# 父类
class Phone:
    IMEI = None #序列号
    producer = "iphone"
    def call_by_4g(self):
        print("4g通话")
# 子类继承父类
class Phone2023(Phone):
    faceId ="1000" # 面部识别id
    def call_by_5g(self):
        print("5g通话")
phone = Phone2023()
phone.call_by_4g()
print(phone.producer)
# 演示多继承
class  NFCReader:
    nfc_type = "第五代"
    producer = "iphone"
    def read_card(self):
        print("NFC读卡")
    def write_card(self):
        print("NFC写卡")
class RemoteControl:
    rc_type = "红外遥控器"
    def control(self):
        print("红外遥控开启了")

class Myphone(Phone,NFCReader,RemoteControl):
    pass
print("++++++++++++++")
my_phone = Myphone()
my_phone.call_by_4g()
my_phone.read_card()
my_phone.write_card()
my_phone.control()
```

```python
"""
    继承里面的复写
    调用父类的同名方法有两种方式：
    方式一：
    使用成员变量:父类名.成员变量
    使用成员方法:父类名.成员方法(self)
    方式二（推荐）：
    使用super()
    使用成员变量: super().成员变量
    使用成员方法: super().成员方法()
"""
"""
演示继承中的复写（方法重写）
"""

# 定义父类
class Phone:
    IMEI = None
    producer = "父类工厂"

    def call(self):
        print("--- 父类方法被调用 ---")
        print("正在使用基础的4G网络进行通话...")

# 定义子类，继承自Phone
class MyPhone(Phone):
    producer = "新一代工厂"  # 复写（重写）父类的成员变量

    # 复写（重写）父类的 call 方法
    def call(self):
        print("--- 子类的复写方法被调用 ---")
        print("开启5G模式...")

        # 在子类中调用父类的同名方法
        
        # 方法一：使用 父类名.方法名(self)
        print("\n>> 方法一：通过'父类名'调用父类方法")
        Phone.call(self)
        
        print("\n>> 方法二：通过'super()'调用父类方法 (推荐)")
        super().call()
        
        print("\n--- 子类方法执行完毕 ---")


# --- 程序开始执行 ---
print("创建一个子类 MyPhone 的对象")
phone = MyPhone()

# 访问被复写的成员变量，会得到子类中定义的值
print(f"手机的制造商是: {phone.producer}")

# 调用方法时，会执行子类中复写后的版本
phone.call()
```

变量的类型注解

    变量的类型注解可以帮助提高代码的可读性和可维护性。
    在Python中，可以使用冒号（:）来指定变量的类型，例如：name: str = "Alice"。
    类型注解不会影响代码的运行，但可以通过静态类型检查工具（如mypy）进行类型检查。

```python
"""
    演示变量的类型注解  
    变量:类型 = 值
"""
import json
import random
from typing import List, Tuple, Dict

#基础数据类型注解
var_1:int = 10
var_2:str = "itheima"
var_3:bool = True
#类对象类型注解
class Student:
    pass
stu:Student = Student()
#基础容器类型注解
my_list:list = [1,2,3]
my_tuple:tuple = (1,2,3)
my_dict:dict = {"itheima":666}
#容器类型详细注解
my_list: List[int] = [1,2,3]
my_tuple: Tuple[int, str, bool] = (1,"itheima",True)
my_dict: Dict[str, int] = {"itheima":666} # 字符串类型键，整数类型值
#在注释中进行类型注解
var_4 = random.randint(1,10) #type:int
var_5 = json.loads('{"name":"zhangsan"}')  #type:dict[str,str]
def func():
    return 10
var_6 = func()  #type:int
#类型注解的限制
var_7:int = "itheima"
var_8:str = 1
```

```python
"""
演示对两数(方法)进行类型注解
"""
#对形参进行类型注解
def add(a:int,b:int):
    return a+b
add(1,2)
#对返回值进行类型注解
def func(data:list) -> list:  # -> 意思是“返回类型为”
    return data
```

```python
"""
    Union类型注解
    Union类型注解用于表示一个变量可以是多种类型中的一种。
    通过使用Union，可以提高代码的灵活性，同时仍然保持类型安全。
    例如，变量可以是整数或字符串，可以使用Union[int, str]进行注解。
"""
"""
演示Union联合类型注解
"""
#使用Union类型，必须先导包
from typing import Union
# 在 Python 3.8 及更早版本中不能使用内置的 list[...] 注解，使用 typing.List 替代
my_list: List[Union[int, str]] = [1, 2, "itheima", "itcast"]
my_dict: Dict[Union[int, str], Union[int, str]] = {
    1: "one",
    "two": 2,
    3: "three",
    "four": 4
}

# 对形参进行Union类型注解
def func(data: Union[int, str]) -> Union[int, str]:  # -> 意思是“返回类型为”
    pass

# 调用时需传入参数
func(1)
```

```python
"""
    演示面向对象的多态特性以及抽象类(接口）的使用
                    父类Animal

        子类Dog        子类Cat         子类.......

                def make_noise(animal:Animal):
                    animal.speak()
    函数方法形参声明接收父类对象
    实际传入子类对象，调用子类重写后的方法

    以父类做定义声明
    以子类做具体实现
    用以获得同一行为，不同状态
"""
# 抽象类:包含抽象方法的类称为抽象类
class Animal:
    # 抽象方法
    def speak(self):
        pass
class Dog(Animal):
    def speak(self):
        print("汪汪汪")
class Cat(Animal):
    def speak(self):
        print("喵喵喵")

def make_noise(animal: Animal):
    """制造噪音，需要传入Animal对象"""
    animal.speak()
#演示多态，使用2个子类对象来调用函数
dog = Dog()
cat = Cat()
make_noise(dog)
make_noise(cat)

print("-----------------")
#演示抽象类
class AC:
    def cool_wind(self):
        """制冷"""
        pass
    def hot_wind(self):
        """制热"""
        pass
    def swing_l_r(self):
        """左右摆风"""
        pass

class Midea_AC(AC):
    def cool_wind(self):
        print("美的制冷")
    def hot_wind(self):
        print("美的制热")
    def swing_l_r(self):
        print("美的左右摆风")
class Gree_AC(AC):
    def cool_wind(self):
        print("格力制冷")
    def hot_wind(self):
        print("格力制热")
    def swing_l_r(self):
        print("格力左右摆风")
def make_cool(ac:AC):
    ac.cool_wind()
gree = Gree_AC()
meidea = Midea_AC()
make_cool(gree)
make_cool(meidea)
```

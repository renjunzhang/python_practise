```python
# 生成1到100之间的随机整数
import random
num = random.randint(1, 100)
count = 0
print(f"这个数字已经生成，是{num}，快来猜吧！")
flag = True
while flag:
    guess = int(input("请输入你猜的数字："))
    count += 1
    if guess > num:
        print("猜大了")
    elif guess < num:
        print("猜小了")
    else:
        print("恭喜你，猜对了！,你一共猜了{}次".format(count))
        flag = False
```

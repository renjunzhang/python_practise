1、while循环

```python
i = 0
while i < 2:
    print("小美我喜欢你")
    i += 1
```

```python
# 生成1到100之间的随机整数
import random
num = random.randint(1, 100)
count = 0
print(f"这个数字已经生成，是{num}，快来猜吧！")
flag = True
while flag:
    guess = int(input("请输入你猜的数字："))
    count += 1
    if guess > num:
        print("猜大了")
    elif guess < num:
        print("猜小了")
    else:
        print("恭喜你，猜对了！,你一共猜了{}次".format(count))
        flag = False
```

```python
"""
    while循环嵌套
    打印99乘法表
    一个外层循环控制行数i
    一个内层循环控制列数j
    end=""表示不换行
"""
i = 1
while i <= 9:
    j = 1
    while j <= i:
        print("{}x{}={}\t".format(j, i, i * j), end="") # end=""表示不换行,\t可以对齐
        j += 1
    print() # 每打印完一行换行
    i += 1
```

2、for循环，也叫遍历循环

    将字符串的内容，逐个取出赋予临时变量，再进行操作。

```python
name = "ladygaga"
for x in name:
    # 将name的内容，逐个取出赋予x临时变量
    # 每次循环打印x的值
    print(x)

"""
    for循环的语法结构：
    for 临时变量 in 可迭代对象：
        循环体
    可迭代对象：字符串、列表、元组、字典、集合、range()函数等
    
    将可迭代对象的内容，逐个取出赋予临时变量，再进行操作。

    for循环的临时变量规范上作用域仅限于循环体内，但在Python中，循环体外也可以访问到该变量。
"""
```

```python
"""
    练习：数一数有多少个'a'
"""
name = "ladygaga"
count = 0
for x in name:
    if x == 'a':
        count += 1
print(f"{name} 中有 {count} 个 'a'")
```

```python
"""
    range()函数的使用
    range(结束值)：生成从0开始，到结束值-1的整数序列
    range(开始值，结束值)：生成从开始值，到结束值-1的整数序列
    range(开始值，结束值，步长)：生成从开始值，到结束值-1，间隔为步长的整数序列
"""
for i in range(5):
    print(i,end=" ")  # 0 1 2 3 4
print()  # 换行
for i in range(3,8):
    print(i,end=" ")  # 3 4 5 6 7
print()  # 换行
for i in range(1,10,2):
    print(i,end=" ")  # 1 3 5 7 9
```

```python
i = 0
for i in range(1,5):
    print(f"Today is day {i} of 4 days of code challenge.")
    for j in range(1,3):
        print(f"  This is message {j} for day {i}.")
print(f"After the loops, i is {i}.")
```

```python
for i in range(1,10):
    for j in range(1,i+1):
        print(f"{j}x{i}={i*j}\t", end="")
    print()
```

3、循环中断

    break：结束整个循环，作用范围是离它最近的循环

    continue：结束本次循环，进入下一次循环，作用范围是离它最近的循环

```python
"""练习：发工资"""
import random
total_salary = 10000
employee = 0
level = 0
for employee in range(1,5):
    level = random.randint(1,10)
    if level <= 5 and total_salary >= 3000:
        print(f"员工{employee}，等级{level}，工资3000元")
        total_salary -= 3000
    elif level > 5 and total_salary >= 5000:
        print(f"员工{employee}，等级{level}，工资5000元")
        total_salary -= 5000
    else:
        print(f"员工{employee}，等级{level}，工资不足，无法发放工资")
print(f"剩余工资总额：{total_salary}元")
```

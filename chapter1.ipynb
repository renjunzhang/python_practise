5、数据类型的转换   int(x), float(x), str(x), bool(x)

```python
"""

    万物都可以转换为字符串
    但是并非所有字符串都可以转换为整数或浮点数

"""

# 将整数和浮点数转换为字符串
num_str = str(123)
print(type(123))
print(num_str, type(num_str))  # 输出: '123' <class 'str'>

float_str = str(45.67)
print(type(45.67))
print(float_str, type(float_str))  # 输出: '45.67' <class 'str'>

# 将字符串转换为整数和浮点数
num = int("123")
print(type("123"))
print(num, type(num))  # 输出: 123 <class 'int'>
float_num = float("45.67")
print(type("45.67"))
print(float_num, type(float_num))  # 输出: 45.67 <class 'float'>

# 将整数转为浮点数
int_to_float = float(100)
print(int_to_float, type(int_to_float))  # 输出: 100.0 <class 'float'>

# 将浮点数转为整数
float_to_int = int(99.99)
print(float_to_int, type(float_to_int))  # 输出: 99 <class 'int'>
```

7、运算符   + - * / // % **  = += -= *= /= //= %= **=  == != > < >= <= 

```python
# 算术运算符示例
print("1 + 1 = ", 1 + 1)  # 输出: 1 + 1 = 2
print("5 - 2 = ", 5 - 2)  # 输出: 5 - 2 = 3
print("3 * 4 = ", 3 * 4)  # 输出: 3 * 4 = 12
print("8 / 2 = ", 8 / 2)  # 输出: 8 / 2 = 4.0
print("7 // 3 = ", 7 // 3)  # 输出: 7 // 3 = 2  整除
print("7 % 3 = ", 7 % 3)  # 输出: 7 % 3 = 1     求余数
print("2 ** 3 = ", 2 ** 3)  # 输出: 2 ** 3 = 8  幂运算

# 赋值运算符示例
a = 10
print("初始值 a =", a)  # 输出: 初始值 a = 10
a += 5
print("a += 5 后 a =", a)  # 输出: a += 5 后 a = 15
a -= 3      
print("a -= 3 后 a =", a)  # 输出: a -= 3 后 a = 12
a *= 2      
print("a *= 2 后 a =", a)  # 输出: a *= 2 后 a = 24
a /= 4      
print("a /= 4 后 a =", a)  # 输出: a /= 4 后 a = 6.0
a //= 2     
print("a //= 2 后 a =", a)  # 输出: a //= 2 后 a = 3.0
a %= 2      
print("a %= 2 后 a =", a)  # 输出: a %= 2 后 a = 1.0
a **= 3
print("a **= 3 后 a =", a)  # 输出: a **= 3 后 a = 1.0
```

1、函数

    函数的定义与调用
    def 函数名(参数列表):
        函数体
        return 返回值
没有返回值的函数默认返回None

```python
"""len()函数的实现"""
def my_len(obj):
    """
    my_len函数用于计算对象的长度
    :param obj: 需要计算长度的对象
    :return: 对象的长度
    """
    count = 0
    for i in obj:
        count += 1
    return count
str1 = "hello"
str2 = [1, 2, 3, 4, 5, 6]

def add(a, b):   # 函数定义时的参数叫做形参
    """
    add函数用于计算两个数的和
    :param a: 第一个加数
    :param b: 第二个加数
    :return: 两个数的和
    """
    print(f"{a} + {b} =", end=" ")
    return a + b


print(my_len(str1))
print(my_len(str2))
print(add(my_len(str1), my_len(str2)))  # 函数调用时的参数叫做实参
```

2、局部变量与全局变量

    局部变量：在函数内部定义的变量，只能在函数内部使用
    全局变量：在函数外部定义的变量，可以在函数内部使用
    使用global关键字可以在函数内部修改全局变量

    全局变量在函数内部被修改时，需要使用global关键字声明该变量为全局变量，否则会创建一个新的同名局部变量

```python
"""练习：ATM"""

def menu():
    """选择操作菜单"""
    print("欢迎使用ATM")
    print("1. 查询余额")
    print("2. 存款")
    print("3. 取款")
    print("4. 退出")
    print("请选择操作（1-4）：", end="")
    return input()

def model(choice):
    """操作功能模块"""
    if choice == "1":
        print("\n----------查询余额----------")
        print("\n您的余额为：1000元")
    elif choice == "2":
        print("----------存款----------")

    elif choice == "3":
        print("----------取款----------")
    elif choice == "4":
        print("----------退出----------")
    else:
        print("输入有误，请重新选择")
    return

model(menu())
```

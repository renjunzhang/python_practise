1、字符串扩展

    单引号定义字符串时，字符串中可以包含双引号
    双引号定义的字符串中可以包含单引号
    三引号可以定义多行字符串

```python
# 单引号定义字符串
name = '小明'
print(type(name))
# 双引号定义字符串
name2 = "小红"
print(type(name2))
# 三引号定义多行字符串
name3 = '''小刚
和
小强'''
print(type(name3))

# 转义字符 \ 解除引号的效用, 让引号作为字符串的一部分
name4 = "He said, \"I'm coming home.\""
print(name4)
```

2、字符串的拼接

    使用 + 号可以将多个字符串拼接成一个字符串

```python
print("大湖名城" + "," + "创新高地")
name = "大湖名城"
slogan = "创新高地"
postcode = "230000" # 不能是整数，只能是字符串
print("合肥是：" + name +"," + slogan + "\n邮编：" + postcode)
```

3、字符串格式化

    使用 % 号可以将变量的值格式化到字符串中

```python
"""
    %s ：
    % 表示，占位
    s 表示，字符串,数字可以转换为字符串
    %d : 整数
    %f : 浮点数
"""
class_num = 57
avg_score = 85.99
message = "我们班有 %s 个同学，平均分是 %s 分。" % (class_num, avg_score)
print(message)
```

4、格式化精度

    使用 %f 可以格式化浮点数，可以指定小数点后的位数

```python
"""
    %5d : 总共占5个字符位，右对齐，左边补空格
    %-5d : 总共占5个字符位，左对齐，右边补空格
    %5.2f : 总共占5个字符位，小数点后保留2位，右对齐，左边补空格
    %.2f : 小数点后保留2位
"""
num1 = 42
num2 = 3.14159
print("整数格式化：'%5d'" % num1)
print("整数格式化：'%-5d'" % num1)
print("整数格式化：'%2d'" % num1) # 如果数字本身位数超过指定宽度，则按实际位数输出
print("浮点数格式化：'%5.2f'" % num2)
print("浮点数格式化：'%.2f'" % num2)
```

5、字符串的格式化2

    使用 format() 方法可以将变量的值格式化到字符串中
    f{占位} 字符串可以将变量的值格式化到字符串中

```python
print(f"The class has {class_num} students, and the average score is {avg_score}.")
```

6、对表达式进行格式化


```python
print("2 * 2 的结果是: %d" % (2 * 2))
print(f"2 * 2 的结果是: {2 * 2}")
print("字符串在Python中的类型是: %s" % type('字符串'))
```

练习

```python
company_name = "Tesla"
stock_price = 699.605
stock_code = "TSLA"
stock_price_daily_growth_factor = 0.025
growth_days = 7
print(f"公司名称: {company_name}, 股票代码: {stock_code}, 当前股价: ${stock_price:.2f}")
print("每日增长系数是: %.3f ，经过 %d 天后，股价将达到: $%.2f" %
      (stock_price_daily_growth_factor, growth_days,
       stock_price * (1 + stock_price_daily_growth_factor) ** growth_days))
```

7、数据输入

    使用 input() 函数可以从控制台获取用户输入的数据
    输入的数据类型是字符串，但是可以将字符串转换为其他数据类型

```python
my_name = input("who you are?\nPlease input your name:")
print(f"Hello, {my_name}!")
age = input("How old are you?\nPlease input your age:")
age = int(age)  # 将输入的字符串转换为整数
print(f"我是 {my_name}，今年 {age} 岁了。")
```

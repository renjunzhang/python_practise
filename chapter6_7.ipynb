数据容器

    列表（list）：有序可变的容器，使用方括号[]定义，支持重复元素，支持嵌套
    元组（tuple）：有序不可变的容器，使用圆括号()定义，支持重复元素
    集合（set）：无序不重复的容器，使用大括号{}或set()函数定义
    字典（dict）：无序的键值对容器，使用大括号{}定义，键必须唯一
    字符串（str）：有序不可变的字符序列，使用单引号''或双引号""定义

1、列表

    列表.append(元素): 在列表末尾追加一个新元素。
    列表.extend(容器): 将另一个容器（如列表、元组）中的所有元素逐个追加到列表末尾。
    列表.insert(下标, 元素): 在指定的下标位置插入一个元素。
    del 列表[下标]: 删除指定下标位置的元素。
    列表.pop(下标): 删除并返回指定下标位置的元素（如果不提供下标，默认删除最后一个）。
    列表.remove(元素): 从列表中删除第一个匹配到的指定元素。
    列表.clear(): 清空列表中的所有元素。
    列表.count(元素): 统计指定元素在列表中出现的次数。
    列表.index(元素): 查找指定元素在列表中的下标，如果找不到会报错 ValueError。
    len(列表): 返回列表中的元素总数。

```python
my_list = ['apple', 'banana', 'cherry', 666, True, [1, 2, 3]]
print(my_list)
print(my_list[0])  # 访问第一个元素
print(my_list[-1]) # 访问最后一个元素
print(my_list[-1][0]) # 访问最后一个元素的第一个元素
print(my_list[1:4]) # 切片操作，获取索引1到3的元素
my_list.append('date')  # 添加元素
print(my_list)
my_list.remove(666)    # 删除元素
print(my_list)
my_list[1] = 'blueberry'  # 修改元素
print(my_list)
print(len(my_list))  # 获取列表长度
print(my_list)
my_list.reverse()   # 反转列表
print(my_list)
print(type(my_list))  # 输出列表的类型
```

4、字符串

    字符串[下标]: 根据下标索引取出特定位置的字符。
    字符串.index(子字符串): 查找给定子字符串在字符串中第一次出现的下标。
    字符串.replace(旧字符串, 新字符串): 将字符串内的所有旧字符串替换为新字符串，返回一个新字符串，原字符串不变。

    len(字符串): 统计字符串中字符的总个数。

5、序列

    序列是指可以通过索引访问其元素的数据类型，包括列表、元组和字符串。
    访问元素：序列[index]，索引从0开始，负数索引表示从末尾开始计数
    切片操作：序列[start:end:step]，获取子序列
    常用函数：len(序列)，max(序列)，min(序列)

```python
#对list进行切片，从1开始，4结束，步长
list = [0,1,2,3,4,5,6]
print(list[1:4:1])
#对tuple进行切片，从头开始，到最后结束，步长1
tuple = (0,1,2,3,4,5,6)
print(tuple[::])
#对str进行切片，从头开始，到最后结束，步长2
str = '0123456'
print(str[::2])
#对str进行切片，从头开始，到最后结束，步长-1
print(str[::-1])
#对列表进行切片，从3开始，到1结束，步长-1
print(list[3:1:-1])
#对元组进行切片，从头开始，到尾结束，步长-2
print(tuple[::-2])
```

```python
my_str='万过薪月，员序程马黑来，nohtyP学'
start = my_str.index('员')
end = my_str.index("黑")
new_str = my_str[start:end+1:]
print(new_str[::-1])
print(new_str[::])
#对比两段代码的输出语句
my_str='万过薪月，员序程马黑来，nohtyP学'
start = my_str.index('员')
end = my_str.index("黑")
print(my_str[start:end + 1:][::-1])
#通过字符串分隔
print(my_str.split("，")[1].replace("来","")[::-1])
```

6、集合（内容无序，不能重复）

    定义格式：set_name = {element1, element2, element3, ...} 或 set_name = set([element1, element2, ...])
    添加元素：set_name.add(element)
    删除元素：set_name.remove(element)（元素不存在会报错），set_name.discard(element)（元素不存在不会报错）
    集合运算：并集（set1 | set2）、交集（set1 & set2）、差集（set1 - set2）、对称差集（set1 ^ set2）
    常用方法：len(), clear(), pop(), update()

```python
# 定义集合
set_1 = {'多','la','A','梦','多'}
# 定义空集合
set_empty = set()
print(f"set_1的内容是{set_1}，类型是{type(set_1)}")
print(f"set_empty的内容是{set_empty}，类型是{type(set_empty)}")
# 添加新元素
set_1.add('python')
print(f"set_1添加后的内容是{set_1}")
# 移除元素
set_1.remove('多')
print(f"set_1移除后的内容是{set_1}")
# 随机取出元素
element= set_1.pop()
print(f"随机取出的元素{element}")
print(f"set_1随机取出后的内容是{set_1}")
# 清空集合
set_1.clear()
print(f"set_1清空后的内容是{set_1}")
# 取出两个集合的差集
set1 = {1,2,3}
set2 = {1,5,6}
set3 = set1.difference(set2)
print(set3)
print(set2)
print(set1)
# 消除两个集合的差集
set1.difference_update(set2)
print(set1)
print(set2)
# 两个集合合并
set1 = {1,2,3}
set2 = {1,5,6}
set3 = set1.union(set2)
print(set3)
print(set2)
print(set1)
# 统计集合元素个数
print(len(set2))
```

    集合.add(元素): 向集合内添加一个元素。
    集合.remove(元素): 移除集合内指定的元素。
    集合.pop(): 从集合中随机取出一个元素并返回。
    集合.clear(): 将集合清空。
    集合1.difference(集合2): 得到一个新集合，包含在集合1中但不在集合2中的元素（差集）。原有的两个集合内容不变。
    集合1.difference_update(集合2): 从集合1中移除所有在集合2中也存在的元素。集合1被修改，集合2不变。
    集合1.union(集合2): 得到一个新集合，包含集合1和集合2的全部元素（并集）。原有的两个集合内容不变。
    len(集合): 统计集合中元素的数量。
可以容纳多个数据；可以容纳不同类型的数据（混装)；数据是无序存储的（不支持下标索引)；不允许重复数据存在；
可以修改（增加或删除元素等)；支持for循环，不支持while循环


```python
my_list=['黑马程序员','传智播客','黑马程序员','传智播客', 'itheima', 'itcast','itheima', 'itcast','best']
my_set = set()
for x in my_list:
    my_set.add(x)
print(my_set)
```

7、字典（key-value 键值对）

    定义格式：dict_name = {key1: value1, key2: value2, ...} 或 dict_name = dict([(key1, value1), (key2, value2), ...])
    访问元素：dict_name[key]
    添加/修改元素：dict_name[key] = value
    删除元素：del dict_name[key]
    常用方法：len(), keys(), values(), items(), get(), pop(), update()
    键值对的Key和Value可以是任意类型(Key不可为字典)
    字典内Key不允许重复，重复添加等同于覆盖原有数据
    字典不可用下标索引，而是通过Key检索Value
    支持for循环，不支持while循环

```python
#定义字典
my_dict1 = {'王':99,'林':88,'周':99}
#定义空字典
my_dict2 = {}
my_dict3 = dict()
print(f"字典1的内容是:{my_dict1}，类型:{type(my_dict1)}")
print(f"字典2的内容是:{my_dict2}，类型:{type(my_dict2)}")
print(f"字典3的内容是:{my_dict3}，类型:{type(my_dict3)}")
#定义重复Key的字典
my_dict4 = {'王':99,'王':88}
print(f"字典4重复key的内容是：{my_dict4}")
#从字典中基于Key获取Value
print(my_dict1['王'])
#定义嵌套字典
subject = {
    '王':{
        '语文':77,
        '数学':66,
        '英语':33
    },'周':{
        '语文': 88,
        '数学': 86,
        '英语': 55
    },'林':{
        '语文': 99,
        '数学': 96,
        '英语': 66
    }
}
# 从嵌套字典中获取数据
print(f"学生的考试信息：{subject}")
print(subject['王']['语文'])
```

    字典.keys(): 获取字典中所有的键 (key)。
    字典.values(): 获取字典中所有的值 (value)。
    字典.items(): 获取字典中所有的键值对 (key-value pair)。
    字典.get(key, [default]): 根据键获取值。如果键不存在，返回 None 或您指定的默认值，不会报错。
    字典.pop(key): 移除指定键的键值对，并返回对应的值。
    字典.update(另一个字典): 用另一个字典的键值对来更新（或添加）当前字典的键值对。
    del 字典[key]: 删除指定的键值对。
    len(字典): 返回字典中键值对的数量。

8、函数多种传递方式

```python
# 位置参数示例
def user_info(name:str,age:int,gender:str): # type: ignore
    print(f"名字是{name}，年龄是{age}，性别是{gender}")
user_info('韩一',18,'男')
```

```python
# 关键字参数示例
def user_info(name ,age,gender):
    print(f"名字是{name}，年龄是{age}，性别是{gender}")
user_info(name='韩二',age=19,gender='男')
user_info("韩三",age=20,gender='女')
```

```python
# 缺省参数示例
def user_info( age,name='null',gender='男'):
    print(f"名字是{name}，年龄是{age}，性别是{gender}")
user_info('韩四',18)
user_info('韩四',18,'女')
```

```python
# 不定长-位置不定长，*号
# 传进的所有参数都会被args变量收集，它会根据传进参数的位置合并为一个元组(tuple)，
# args是元组类型，这就是位置传递
def user_info(*args):
    print(f"args参数的类型是{type(args)},内容是{args}")
user_info('小李',18)
# 不定长-关键字不定长，**号
# 参数是“键=值”形式的形式的情况下，所有的“键=值”都会被kwargs接受，
# 同时会根据“键=值”组成字典.
def user_info(**kwargs):
    print(f"args参数的类型是{type(kwargs)},内容是{kwargs}")
user_info(name='小李',age = 18)
```

```python
"""
    lambda匿名函数示例
    这是无名函数，只可以临时使用,只可以写一行代码
"""
# 定义一个函数，接受其它函数输入
def test_function(compute):
    result = compute(1,2)
    print(f"结果是{result}")
# 通过lambda匿名函数的形式，将匿名函数作为参数传入
test_function(lambda x,y: x*y) # 参数是x和y，返回值是x*y

```
